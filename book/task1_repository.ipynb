## Approach and reasoning

Command: `bash scripts/run_stage2_evidence.sh`. The runbook shallow-clones the dataset repository, resolves the real 40-character commit SHA, and records the selected Python files from `src/` and `tests/`. The notebook reads the captured manifest instead of needing a live clone during publication.

## What this chapter proves

| Requirement | Evidence in this chapter |
|---|---|
| Repository discovery | The executed cell prints total Python files and selected `src`/`tests` files. |
| Dataset provenance | The output includes the pinned dataset commit SHA. |
| Reproducible publication | The Jupyter Book build can render the saved output without network or service access. |


````python
from pathlib import Path
import json, subprocess
ROOT = next(p for p in [Path.cwd(), *Path.cwd().parents] if (p / 'screenshots/stage2_manifest.json').exists())
repo = ROOT / 'data/datasets'
manifest = json.loads((ROOT / 'screenshots/stage2_manifest.json').read_text())
all_python = sorted(repo.rglob('*.py'))
selected = [p for p in all_python if any(part in {'src', 'tests'} for part in p.relative_to(repo).parts)]
print('dataset_commit:', subprocess.check_output(['git','-C',str(repo),'rev-parse','HEAD'], text=True).strip())
print('all_python_files:', len(all_python))
print('selected_src_tests_files:', len(selected))
print('sample:', [str(p.relative_to(repo)) for p in selected[:5]])
````
